# Do Geometry Stuff With The `geom` library

```python
import sys
sys.path.insert(0, '../')

from src import words
from src import curves
from src import geom

import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
```

```python
# define a custom word-plotter real quick

def plot_curve(curve, grid=True, show_ep=True, title=None, show=True):
    x,y = curve.T
    xs, ys = curve[0]
    xf,yf = curve[-1]
    plt.gca().set_aspect('equal')
    if grid:
        plt.grid()
    plt.plot(x,y, c='black')
    if show_ep:
        plt.plot([xs],[ys],c='green',marker='>')
        plt.plot([xf],[yf],c='red',marker='s')
    if not title is None:
        plt.title(title)
    if show:
        plt.show()
```

```python
alpha = 2 * np.pi  / 5
myword = words.knacci(9,2)
curve = curves.get_curve(myword, alpha)

plot_curve(curve, title="Look at those intersections!")
```

```python
these_intersects = geom.intersects(curve)
for intersection in these_intersects:
    print(intersection)
```

```python
for idx, inter in enumerate(these_intersects):
    subcurve = curve[inter['n']:inter['m']]
    plot_curve(subcurve)
```

```python
alpha = 2 * np.pi  / 5
myword = words.knacci(7,2)
curve = curves.get_normed_curve(myword, alpha)

hull = ConvexHull(curve)
x,y = curve.T
plt.gca().set_aspect('equal')
plt.plot(x,y)
for simplex in hull.simplices:
    plt.plot(curve[simplex, 0], curve[simplex, 1], 'k-')
    
s_7 = [curve[simplex] for simplex in hull.simplices]
```
